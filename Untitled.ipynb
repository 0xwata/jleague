```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://ultra.zone/national-football-team-japan')
response.encoding = response.apparent_encoding  # この行を追加
soup = BeautifulSoup(response.text, 'html.parser') # BeautifulSoupの初期化
print(soup.prettify()) # HTMLをインデントすることができます。
```

```python
texts_1 = soup.find_all('script')
texts_2 = soup.find('script')
```

```python
objs = texts_1[8].string.split('\r\n\t')
player_name_list = []
for obj in objs:
    if re.search('\[.*?\]',obj):
        match_players_names = re.search('\[.*?\]',obj).group().strip('[]').split(',')
        for  player_name in match_players_names :
            player_name = player_name.strip('""')
            if player_name == '':
                pass
            else:
                player_name_list.append(player_name)
player_name_list = list(set( player_name_list)) 
player_name_list = pd.DataFrame({'player_name': player_name_list})
```

```python
player_name_list.to_csv('player_name.csv')
```
