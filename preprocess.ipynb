# 不快指数計算 -> テストデータに欠損値ありで使えない。

```python
# # print(whole[['id','humidity']])
whole['temperature'] = whole['temperature'].fillna(0)
whole['humidity'] = whole['humidity'].fillna(0)
whole['DI'] = (whole['temperature'] * 0.81 + 0.01*whole['humidity']*(0.99*whole['temperature']-14.3) + 46.3).astype('i')

def grouping_displeasure(val):
    if val <= 55: #寒い
        val = 0
    elif 55 < val and val <= 60: #肌寒い
         val = 1
    elif 60 < val and val <= 65: #何も感じない
         val = 1
    elif 65 < val and val <= 70: #快い
         val = 1
    elif 70 < val and val <= 75: #暑くない
         val = 1
    elif 75 < val and val <= 80: #やや暑い
         val = 0
    elif 80 < val and val <= 85:#暑くて汗が出る
         val = 0
    elif val >= 85: #暑くてたまらない
        val = 0
    return val 

whole['DI'] = whole['DI'].apply(grouping_displeasure)

whole['temperature'] =(((whole['temperature'].fillna(0))/5).astype('i'))*5
whole['humidity'] =(((whole['humidity'].fillna(0))/10).astype('i'))*10
whole['humidity'] = whole['humidity'].apply(lambda x: 1 if x <= 70 and x >= 40 else 0)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.countplot(whole['DI'])

plt.subplot(1, 2, 2)
sns.boxplot(x='DI', y='y' , data=whole)


plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.countplot(whole['humidity'])

plt.subplot(1, 2, 2)
sns.boxplot(x='humidity', y='y' , data=whole)



plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.countplot(whole['temperature'])

plt.subplot(1, 2, 2)
sns.boxplot(x='temperature', y='y' , data=whole)

```
